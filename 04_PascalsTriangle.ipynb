```python
dictFact = {}
def factorial(n):
    global dictFact
    if n in dictFact:
        return dictFact[n]
    if n <= 1:
        return 1
    else:
        dictFact[n] = n * factorial(n-1)
        return n * factorial(n-1)
```

```python

for i in range(0,99):
    z = 0
    for j in range(0,i+1):
        z = int(factorial(i) /( factorial(j) * factorial(i - j)))
        print(z, end=' ')
    print()
```
